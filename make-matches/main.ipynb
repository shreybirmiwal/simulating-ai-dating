```python
import pandas as pd


df = pd.read_csv('data.csv')
```

```python
# delete every row that has  'could not extract' in the row some text

df = df[~df.apply(lambda row: row.astype(str).str.contains('Could not extract').any(), axis=1)]
```

```python
df
```

```python
all_matches = []  # Global list to track matched pairs (stored as strings like "rowA:rowB")

def checkLocation(row1, row2):
    if pd.isna(row1['LocationFlexibility']) or pd.isna(row2['LocationFlexibility']):
        loc1 = [x.strip() for x in str(row1['Location']).split(',')]
        loc2 = [x.strip() for x in str(row2['Location']).split(',')]

        #print(loc1, loc2, "BRO")

        for locs in loc1:
            if locs in loc2:
                return True
    else:
        return True

    return False

def get_matches(row):
    my_gen = row['Gender'].strip()
    my_want = [x.strip() for x in str(row['InterestedIn']).split(',')]
    
    matches = []  # Initialize the matches list for this row
    
    for idx, other_row in df.iterrows():
        if idx == row.name:  # Skip self
            continue
        
        # Check if other's gender is in my desired list
        if other_row['Gender'].strip() in my_want:
            # Check if my gender is in their desired list
            other_want = [x.strip() for x in str(other_row['InterestedIn']).split(',')]
            if my_gen in other_want:
                pair1 = f"{row.name}:{other_row.name}"
                

                if checkLocation(row, other_row):

                    matches.append(other_row.name) 
                    all_matches.append(pair1) 


                    
    return matches  # Return the matches list (not the entire row)

# Apply the function and assign to 'matches' column
df['matches'] = df.apply(get_matches, axis=1)
```
